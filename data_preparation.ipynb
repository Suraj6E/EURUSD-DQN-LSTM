```python
# Plot the time series data
import pandas as pd
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objects as go

filename = "EURUSD_M1_2023.csv"

data = pd.read_csv("./data_original/"+filename)

print(data.head(10))
df = pd.DataFrame()
df["timestamp"] = pd.to_datetime(data.iloc[:, 0] +" "+data.iloc[:, 1] + ":00");
df["open"] = data.iloc[:, 2]
df["high"] = data.iloc[:, 3]
df["low"] = data.iloc[:, 4]
df["close"] = data.iloc[:, 4]

df.set_index('timestamp', inplace=True)
df_resampled = df.resample('D').agg({'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last'}).reset_index()

fig = go.Figure(data = [go.Candlestick(x = df_resampled['timestamp'],
                                       open = df_resampled['open'],
                                       high = df_resampled['high'],
                                       low = df_resampled['low'],
                                       close = df_resampled['close'])])

fig.show()

## save data
# df.to_csv('./data_saved/'+filename)
```
